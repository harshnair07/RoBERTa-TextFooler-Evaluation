```python
import pandas as pd
from datasets import load_dataset

print("1. Loading HC3 Dataset (Default Config)...")
# We remove name="all" and use the standard load method
# Using revision="refs/convert/parquet" is still recommended for speed in the lab
dataset = load_dataset("Hello-SimpleAI/HC3", revision="refs/convert/parquet", split="train")

# 2. Convert to Pandas
# We'll take the first 1,000 rows to ensure we have a large, diverse set
df = dataset.to_pandas().head(1000)

attack_data = []

print("2. Processing 2,000 samples (1,000 Human + 1,000 AI)...")
for _, row in df.iterrows():
    # Extract strings from the lists (HC3 stores answers in lists)
    if len(row['human_answers']) > 0 and len(row['chatgpt_answers']) > 0:
        # Label 0 = Human
        attack_data.append({"text": row['human_answers'], "label": 0})
        # Label 1 = AI
        attack_data.append({"text": row['chatgpt_answers'], "label": 1})

# 3. Shuffle the data so the model doesn't see patterns in the order
final_df = pd.DataFrame(attack_data).sample(frac=1).reset_index(drop=True)

# 4. Save to CSV
final_df.to_csv("adversarial_test_data_2000.csv", index=False)
print(f"✅ Success! Created 'adversarial_test_data_2000.csv' with {len(final_df)} samples.")
```
